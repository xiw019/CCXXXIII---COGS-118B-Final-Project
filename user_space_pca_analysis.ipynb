```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Read data
results_path = './results'
user_space_pca_res = np.load(os.path.join(results_path, 'user_space_pca_res.npz'))
```

```python
user_id_counts = user_space_pca_res['arr_0']
pca_var_rate = user_space_pca_res['arr_1']
mse_all = user_space_pca_res['arr_2']
rmse_all = user_space_pca_res['arr_3']
```

```python
print('Number of users with >= 1000 ratings: ', len(rmse_all[user_id_counts >= 1000]))
print('Number of users with >= 500 ratings: ', len(rmse_all[user_id_counts >= 500]))
print('Number of users with >= 100 ratings: ', len(rmse_all[user_id_counts >= 100]))
print('Number of users with >= 50 ratings: ', len(rmse_all[user_id_counts >= 50]))
print('Number of users with >= 10 ratings: ', len(rmse_all[user_id_counts >= 10]))

print('Average RMSE over users with >= 1000 ratings: ', np.mean(rmse_all[user_id_counts >= 1000]))
print('Average RMSE over users with >= 500 ratings: ', np.mean(rmse_all[user_id_counts >= 500]))
print('Average RMSE over users with >= 100 ratings: ', np.mean(rmse_all[user_id_counts >= 100]))
print('Average RMSE over users with >= 50 ratings: ', np.mean(rmse_all[user_id_counts >= 50]))
print('Average RMSE over users with >= 10 ratings: ', np.mean(rmse_all[user_id_counts >= 10]))
```

```python
plt.plot(user_id_counts, pca_var_rate)
plt.xlabel('Number of ratings made by a user')
plt.ylabel('PCA-explained variance ratio')
```

```python
plt.plot(user_id_counts, mse_all)
plt.xlabel('Number of ratings made by a user')
plt.ylabel('MSE')
```

```python
plt.plot(user_id_counts, rmse_all)
plt.xlabel('Number of ratings made by a user')
plt.ylabel('RMSE')
```
